### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = r"..\PyBer_Analysis\city_data.csv"
ride_data_to_load = r"..\PyBer_Analysis\ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
ride_total = pyber_data_df.groupby("type").count()["ride_id"]
ride_total
```

```python
# 2. Get the total drivers for each city type
driver_total = pyber_data_df[["city","type","driver_count"]].drop_duplicates().groupby('type')['driver_count'].sum()
driver_total
```

```python
#  3. Get the total amount of fares for each city type
total_fare_city = pyber_data_df.groupby('type')['fare'].sum()
total_fare_city
```

```python
#  4. Get the average fare per ride for each city type. 
avg_fare_city = pyber_data_df.groupby('type')['fare'].sum()/pyber_data_df.groupby('type')['ride_id'].count()
avg_fare_city
```

```python
# 5. Get the average fare per driver for each city type. 
avg_fare_driver = total_fare_city/driver_total
avg_fare_driver
```

```python
#  6. Create a PyBer summary DataFrame. 
pyber_summary_df = pd.DataFrame([ride_total,driver_total,total_fare_city,avg_fare_city,avg_fare_driver]).transpose()
pyber_summary_df = pyber_summary_df.rename({'ride_id':'Total Rides', 'driver_count':'Total Drivers', 'fare':'Total Fares', 'Unnamed 0':'Average Fare per Ride', 'Unnamed 1':'Average Fare per Driver'}, axis='columns')
pyber_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
```

```python
#  8. Format the columns.
formatted_pyber_summary = pyber_summary_df.style.format({
    'Total Rides':'{:,.0f}'.format,
    'Total Drivers':'{:,.0f}'.format,
    'Total Fares':'${:,.2f}'.format,
    'Average Fare per Ride':'${:,.2f}'.format,
    'Average Fare per Driver':'${:,.2f}'.format
})
formatted_pyber_summary
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
import numpy as np
from matplotlib.dates import DateFormatter, MonthLocator

# 1. Read the merged DataFrame
city_data_to_load = r"..\PyBer_Analysis\city_data.csv"
ride_data_to_load = r"..\PyBer_Analysis\ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

pyber_data_df.head()
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
df = pyber_data_df.groupby(["date","type"])["fare"].sum()
df
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
df = df.reset_index()
df
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
pivot_df = pd.pivot_table(df, index = 'date', columns = 'type', values='fare')
pivot_df
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
pivot_df = pivot_df.loc['2019-01-01':'2019-04-28']
pivot_df
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
pivot_df.index = pd.to_datetime(pivot_df.index)
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
pivot_df.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
resampled_df = pivot_df.resample('W').sum()
resampled_df
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')

fig, ax = plt.subplots(figsize=(17,5))
ruralplot = ax.plot(resampled_df["Rural"],label = "Rural")
suburbanplot = ax.plot(resampled_df["Suburban"],label = "Suburban")
urbanplot = ax.plot(resampled_df["Urban"],label = "Urban")

#format the graph
ax.set_title("Total Fare by City Type")
ax.set_ylabel("Fare ($USD)")
ax.set_xlabel("2019",loc = "left", fontsize = "14")
ax.xaxis.set_major_locator(MonthLocator(bymonth=(1,2,3,4)))
ax.xaxis.set_major_formatter(DateFormatter('%b'))

#make a legend
legend = ax.legend(loc = "center", title = "type")

#save the graph
plt.savefig("../PyBer_Analysis/PyBer_fare_summary.png")
```
